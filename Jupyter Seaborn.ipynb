```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('dataset.csv')
display(df)
df2 = pd.read_csv('dataset2.csv')
display(df2)


#df == loja
#df2 == plantas

#Essa biblioteca é uma das mais populares para visualização de dados e oferece diversas ferramentas para criar gráficos elegantes e informativos. Seaborn é uma biblioteca de visualização de dados em Python baseada em Matplotlib.

#Se seaborn é construído tendo como base matplotlib, por que não aprender apenas matplotlib?

#Bom porque ele ele usa belos temas para decorar gráficos Matplotlib :D.
```

```python
# Método é usado para criar gráficos de dispersão. Ele permite visualizar a relação entre duas variáveis numéricas, com os pontos de dados representados no gráfico.

sns.scatterplot(x='Valor_Venda', y='Quantidade', data=df)
plt.show()
```

```python
# Método Lineplot: é usado para criar gráficos de linhas. É útil para visualizar tendências ao longo do tempo ou em relação a uma variável categórica.


sns.lineplot(x='petal_length', y='petal_width', data=df2)
plt.show()
```

```python
# Método Countplot:  é usado para criar gráficos de contagem de variáveis categóricas. Ele mostra quantas vezes cada categoria aparece em seus dados.

sns.countplot(x='Quantidade', data=df)
plt.show()
```

```python
# Método displot: é usado para criar histogramas e estimativas de densidade de probabilidade de uma variável numérica. O histograma é uma maneira fácil de visualizar a distribuição de uma variável.



sns.displot(df['Quantidade'])
plt.show()
```
