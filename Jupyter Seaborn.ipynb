```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('dataset.csv')
display(df)
df2 = pd.read_csv('dataset2.csv')
display(df2)


#df == loja
#df2 == plantas

#Essa biblioteca Ã© uma das mais populares para visualizaÃ§Ã£o de dados e oferece diversas ferramentas para criar grÃ¡ficos elegantes e informativos. Seaborn Ã© uma biblioteca de visualizaÃ§Ã£o de dados em Python baseada em Matplotlib.

#Se seaborn Ã© construÃ­do tendo como base matplotlib, por que nÃ£o aprender apenas matplotlib?

#Bom porque ele ele usa belos temas para decorar grÃ¡ficos Matplotlib :D.
```

```python
# MÃ©todo Ã© usado para criar grÃ¡ficos de dispersÃ£o. Ele permite visualizar a relaÃ§Ã£o entre duas variÃ¡veis numÃ©ricas, com os pontos de dados representados no grÃ¡fico.

sns.scatterplot(x='Valor_Venda', y='Quantidade', data=df)
plt.show()
```

```python
# MÃ©todo Lineplot: Ã© usado para criar grÃ¡ficos de linhas. Ã‰ Ãºtil para visualizar tendÃªncias ao longo do tempo ou em relaÃ§Ã£o a uma variÃ¡vel categÃ³rica.


sns.lineplot(x='petal_length', y='petal_width', data=df2)
plt.show()
```

```python
# MÃ©todo Countplot:  Ã© usado para criar grÃ¡ficos de contagem de variÃ¡veis categÃ³ricas. Ele mostra quantas vezes cada categoria aparece em seus dados.

sns.countplot(x='Quantidade', data=df)
plt.show()
```

```python
# MÃ©todo displot: Ã© usado para criar histogramas e estimativas de densidade de probabilidade de uma variÃ¡vel numÃ©rica. O histograma Ã© uma maneira fÃ¡cil de visualizar a distribuiÃ§Ã£o de uma variÃ¡vel.



sns.displot(df['Quantidade'])
plt.show()
```
